```python
import cv2
import os
for _ in range(1,13):
    video_filename = f'{_}.mp4' #확장자 명을 포함한 동영상 파일을 입력 하기
    output_dir = f'./15프레임/{_}' #이건 폴더명이 동영상 명으로 바뀌게끔 수정하기

    # 스크린샷 저장 디렉토리 생성
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 영상 파일 열기
    cap = cv2.VideoCapture(video_filename)

    if not cap.isOpened():
        print("Could not open video file")
        exit()

    # 프레임 레이트 가져오기
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    print(f'FPS: {fps}')

    # 총 프레임 수 가져오기
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f'Total Frame Count: {frame_count}')

    # 스크린샷 간격 설정 
    interval = fps // 12.5 if fps >= 15 else 1   # FPS가 충분히 크지 않으면 interval은 최소값인 '1'로 설정됩니다.
    print(f'Screenshot Interval: {interval}')

    # 스크린샷 찍기
    frame_number = 0
    screenshot_count = 0
    i = 0
    while True:



        ret, frame = cap.read()

        if not ret:
            print("Could not read frame")
            break

        if frame_number % interval == 0:
            screenshot_filename = os.path.join(output_dir, f'screenshot_{i}.jpg')
            cv2.imwrite(screenshot_filename, frame)
            screenshot_count += 1
            print(f'Saved screenshot at frame {i}')
            i += 1

        frame_number += 1

        if frame_number >= frame_count:
            break

    # 영상 파일 닫기
    cap.release()

    print(f'Saved {screenshot_count} screenshots.')
```
